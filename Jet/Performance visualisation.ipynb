```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
def plot_metrics(df, columns, nrow, ncol):
    fig, axes = plt.subplots(nrow, ncol)
    i = 0
    for r in range(nrow):
        for c in range(ncol):
            df[columns[i]].plot(ax=axes[r,c], title=columns[i])
            i += 1
```

```python
algo = 'a2c'
df = pd.read_csv(f"sb-{algo}/progress.csv")
columns = ['policy_entropy', 'explained_variance', 'value_loss', 'fps']
plot_metrics(df, columns, nrow=2, ncol=2)
```

```python
algo = 'acktr'
df = pd.read_csv(f"sb-{algo}/progress.csv")
df.head()
columns = ['value_loss', 'policy_entropy', 'explained_variance', 'policy_loss']
plot_metrics(df, columns, nrow=2, ncol=2)
```

```python
algo = 'trpo'
df = pd.read_csv(f"sb-{algo}/progress.csv")
df.head()
columns = ['optimgain', 'entropy', 'meankl', 'EpLenMean', 'surrgain', 'EpRewMean', 'explained_variance_tdlam_before', 'entloss']
plot_metrics(df, columns, nrow=2, ncol=4)
```

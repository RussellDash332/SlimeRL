```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
plt.style.use("seaborn")

df = pd.read_csv("sb-a2c/progress.csv")
display(df.head())
display(df.tail())
```

```python
columns = ['policy_entropy', 'explained_variance', 'value_loss', 'fps']

nrow = 2
ncol = 2
fig, axes = plt.subplots(nrow, ncol)

i = 0
for r in range(nrow):
    for c in range(ncol):
        df[columns[i]].plot(ax=axes[r,c], title=columns[i])
        i += 1
```
